```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
```

```python
df = pd.read_csv('stats.csv')
```

```python
df.head()
```

```python
df['violence_per_hundred_thousand'] = (df['count'] / df['population']) * 100000
```

```python
df.head()
```

```python
country_df = df.drop_duplicates(subset=['country'])

c0 = country_df[country_df['score'] == 0]['country']
c1 = country_df[country_df['score'] == 1]['country']
c2 = country_df[country_df['score'] == 2]['country']

print(len(c0), len(c1), len(c2))
```

```python
c0_samples = list(c0.sample(n=5))
c1_samples = list(c1.sample(n=5))
c2_samples = list(c2.sample(n=5))
```

```python
samples = [c0_samples, c1_samples, c2_samples]

for index, countries in enumerate(samples):
  freedom = ""
  if index == 0: freedom = "Not Free"
  elif index == 1: freedom = "Partly Free"
  elif index == 2: freedom = "Free"

  fig = make_subplots(rows=2, cols=3, shared_yaxes=True)
  grid = [[1, 1], [1, 2], [1, 3], [2, 1], [2, 2]]
 
  for index, item in enumerate(grid):
    fig.add_trace(
        go.Scatter(x=df[df['country'] == countries[index]]['year'].values, y=df[df['country'] == countries[index]]['violence_per_hundred_thousand'].values, name=countries[index]),
        row=item[0], col=item[1] 
    )

  fig.update_layout(height=600, width=800, title_text=f"Violence against Citizens across the Years in '{freedom} countries'")
  fig.update_xaxes(type='category')

  fig.show()
```
